<p>Perhaps a histogram may not be the most suitable, lets try a density plot instead. </p>

```python
price = importantBtcData["price(USD)"]

plt.xlabel("Prices (USD)")
plt.title("Distribution of prices")
# plt.hist(price)
sb.set_style('whitegrid')
sb.kdeplot(price)

price.skew()
```

```python
logPrice = np.log(importantBtcData['price(USD)'])


plt.xlabel("Log(Price) in USD")
plt.title("Distribution of Log(Prices)")
sb.set_style('whitegrid')
sb.kdeplot(logPrice)
logPrice.skew()
```

```python
print("Skew of price vs Log(Price):", price.skew(), "vs", logPrice.skew())
print("As seen from the 2 graphs above, Distribution of Prices vs Distribution of LogPrices, the graph became less negative skewed")
```

```python
averageDifficulty = importantBtcData["averageDifficulty"]

plt.xlabel("averageDifficulty")
plt.title("Distribution of averageDifficulty")
# plt.hist(price)
sb.set_style('whitegrid')
sb.kdeplot(averageDifficulty)

averageDifficulty.skew()
```

```python
logAverageDifficulty = np.log(importantBtcData['averageDifficulty'])


plt.xlabel("Log(averageDifficulty)")
plt.title("Distribution of Log(averageDifficulty)")
sb.set_style('whitegrid')
sb.kdeplot(logAverageDifficulty)
logAverageDifficulty.skew()
```

```python
activeAddresses = importantBtcData["activeAddresses"]

plt.xlabel("activeAddresses")
plt.title("Distribution of activeAddresses")
# plt.hist(price)
sb.set_style('whitegrid')
sb.kdeplot(activeAddresses)

activeAddresses.skew()
```

```python
logActiveAddresses = np.log(importantBtcData['activeAddresses'])


plt.xlabel("Log(ActiveAddresses)")
plt.title("Distribution of Log(ActiveAddresses)")
sb.set_style('whitegrid')
sb.kdeplot(logActiveAddresses)
logActiveAddresses.skew()
```

```python
print("Skew of Average Difficulty vs Log(Average Difficulty):", activeAddresses.skew(), "vs", logActiveAddresses.skew())
print("Here, the graph instead becomes more skewed, perhaps the data from active addresses do not follow a log-normal distribution, thus resulting in such. ")
```

```python
dailyHashRate = importantBtcData["Daily Hash Rate(Trillions of hashes/s)"]

plt.xlabel("Daily Hash Rate(Trillions of hashes/s)")
plt.title("Distribution of Daily Hash Rate(Trillions of hashes/s)")
# plt.hist(price)
sb.set_style('whitegrid')
sb.kdeplot(dailyHashRate)

dailyHashRate.skew()
```

```python
logDailyHashRate = np.log(importantBtcData['Daily Hash Rate(Trillions of hashes/s)'])


plt.xlabel("Log(Daily Hash Rate)")
plt.title("Distribution of Log(Daily Hash Rate)")
sb.set_style('whitegrid')
sb.kdeplot(logDailyHashRate)
logDailyHashRate.skew()
```

```python
print("Skew of Daily Hash Rate vs Log(Daily Hash Rate):", dailyHashRate.skew(), "vs", logDailyHashRate.skew())
print("As seen from the 2 graphs above, Distribution of Daily Hash Rate vs Distribution of Log(Daily Hash Rate), the graph became less skewed, albeit in another direction")
```
