<p> It appears that there are some missing dates and data. The number of rows of data under bitcoin hashrate is lesser than that of the filtered Bitcoin data. Hence, we have to remove the extra or fill in the missing dates. Here, we opt to delete the missing data since we are unclear of how the relationship is like before injecting artificial data</p>

```python
filteredDateBitcoin_data.date.tail(30)
```

```python
filteredDateBitcoin_data=filteredDateBitcoin_data.sort_values(by='date')
filteredDateBitcoin_data
```

<p>Here, we have several options when dealing with missing data.</p>

```python
filteredDateBitcoin_data = filteredDateBitcoin_data[(filteredDateBitcoin_data.date <="2018-10-31")]
```

<p> we can use their intersection dates in order to combine the data sets together </p>

```python
final_data = pd.merge(left = filteredDateBitcoin_data, left_on = "date",
        right = bitcoin_hashrate, right_on = "Date")
final_data.drop(columns = ["Date"])
final_data = final_data.rename(columns = {"Value": "Hashrate"})
final_data
```

```python
final_data.shape #2010 rows and 19 columns 
```

<p> Here is the final data we can use to observe and discover the relationship between each of them </p>

```python
final_data.describe()
```

<p>Using the cleaned dataset, we can now accurately find the relationship between all variables using a heatmap.</p>

```python
f = plt.figure(figsize=(24, 12))
sb.heatmap(final_data.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 20}, cmap = "RdBu")
```

```python
finalizedBitcoin_data = pd.DataFrame(final_data[["price(USD)", "averageDifficulty","cumulative total number of coins","activeAddresses", "Hashrate"]])
```

<p>Let's see all of our finalised data using a pairplot and a heatmap </p>

```python
sb.pairplot(data = finalizedBitcoin_data)
```

```python
f = plt.figure(figsize=(24, 12))
sb.heatmap(finalizedBitcoin_data.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 20}, cmap = "RdBu")
```
